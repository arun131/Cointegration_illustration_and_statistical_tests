```python
import numpy as np
import pandas as pd

def normalize_values(data):
    """
    Normalize the values of a numpy array, list, or pandas Series between 0 and 1.

    Args:
    - data (numpy.ndarray, list, pandas.Series): Input data to be normalized.

    Returns:
    - numpy.ndarray, pandas.Series: Normalized data.
    """

    # Convert data to numpy array if it's not already
    if not isinstance(data, np.ndarray):
        data = np.array(data)

    # Calculate the minimum and maximum values
    min_val = np.min(data)
    max_val = np.max(data)

    # Normalize the data between 0 and 1
    normalized_data = (data - min_val) / (max_val - min_val)

    # If input is pandas Series, return a Series with the same index
    if isinstance(data, pd.Series):
        normalized_data = pd.Series(normalized_data, index=data.index)

    return normalized_data
```

# 2. Mean Reversion Tests

```python
import matplotlib.pyplot as plt

from mean_reversion_tests import hurst_exponent_test, ornstein_uhlenbeck_test, mean_reverting_half_life, engle_granger_test
```

### 2.1 Generate data and conduct tests

```python
asset_data = []
spread_data = []
test_results = []
for i in range(10):
  results = {}
  choice = np.random.choice([1,2,3])
  if choice == 1:
    x, y = not_cointegrated()
  elif choice == 2:
    x, y = simple_method()
  else:
    x, y = adapted_method()

  asset_data.append([x,y])

  beta, alpha = np.polyfit(x, y, 1)
  spread = (beta * x + alpha) - y

  spread_data.append(spread)

  results['hurst'] = hurst_exponent_test(spread)
  results['ou'] = ornstein_uhlenbeck_test(spread)[0]
  results['half_life'] = mean_reverting_half_life(spread)
  results['eg_res'] = engle_granger_test(spread)

  test_results.append(results)
```

```python
test_names = ['hurst', 'ou', 'half_life', 'eg_res']

# Sort test results by each result
sorted_test_results = {}
for test_name in test_names:
    sorted_test_results[test_name] = sorted(range(len(test_results)), key=lambda x: test_results[x][test_name])
```

```python
def plot_test_results(test_name):
  order = sorted_test_results[test_name]
  fig, axs = plt.subplots(10, 2, figsize=[10, 15])
  axst = [ax.twinx() for ax in axs]
  for i, idx in enumerate(order):
      if i%2 == 0:
        x, y = asset_data[idx]
        axs[i].plot(x)
        axs[i].set_ylabel(f'Asset A')
        axst[i].plot(y)
        axst[i].set_ylabel(f'Asset B')

      if i%2 == 1:
        spread = spread_data[idx]
        std = np.std(spread)
        mean = np.mean(spread)
        axs[i].plot(spread)
        axs[i].plot([mean - 2*std] * len(spread))
        axs[i].plot([mean] * len(spread))
        axs[i].plot([mean + 2*std] * len(spread))
        axs[i].set_ylabel(f'{test_name} : {test_results[idx][test_name]:.2f}')
      if i < len(order) - 2:
          axs[i].set_xticks([])

  plt.suptitle(f'Spread Data sorted by {test_name}', y =0.9)
  # plt.tight_layout()
  plt.show()
```

```python
def plot_test_results(test_name):
    order = sorted_test_results[test_name]
    fig, axs = plt.subplots(10, 2, figsize=[15, 15])
    axst = [ax.twinx() for ax in axs.flatten()]
    for i, idx in enumerate(np.repeat(order,2)):
        if i % 2 == 0:
            x, y = asset_data[idx]
            axs[i//2, 0].plot(x)
            axs[i//2, 0].set_ylabel('Asset A')
            axst[i].plot(y, color='orange')
            axst[i].set_ylabel('Asset B')

        if i % 2 == 1:
            spread = spread_data[idx]
            std = np.std(spread)
            mean = np.mean(spread)
            axs[i//2, 1].plot(spread)
            axs[i//2, 1].plot([mean - 2*std] * len(spread), linestyle='--', color='gray')
            axs[i//2, 1].plot([mean] * len(spread), linestyle='--', color='gray')
            axs[i//2, 1].plot([mean + 2*std] * len(spread), linestyle='--', color='gray')
            axs[i//2, 1].set_ylabel(f'{test_name} : {test_results[idx][test_name]:.2f}')
        if i < len(order) - 2:
            axs[i//2, 1].set_xticks([])

    plt.suptitle(f'Spread Data sorted by {test_name}')#, y=0.95)
    plt.tight_layout()
    plt.show()
```

### 2.2 Hurst component for mean reversion

The lower husrt component shows higher mean reverting behaviour.

    - h < 0.5 -- The time series is mean reverting
    - h = 0.5 -- The time series is a Geometric Brownian Motion
    - h > 0.5 -- The time series is trending

```python
plot_test_results("hurst")
```

### 2.3 Ornstein-Uhlenbeck process for mean reversion

    The OU process test gives us a mean reversion speed and long term mean, the higher the mean reversion speed, the better it is at mean reversion

```python
plot_test_results("ou")
```

### 2.4 Mean Reversion Half life

      Lower the half life estimate, the better it is at mean reversion.

```python
plot_test_results('half_life')
```

### 2.5 Engle-Granger Test for Mean Reversion

    Engle-Granger two-step cointegration test first checks for co-integration and then tests for stationarity. If both tests are satisfied, we say that a pair is mean-reverting.


```python
plot_test_results("eg_res")
```

### 2.6 Summary

Some tests are very useful like the hurst exponent and the half life estimate. We can use these tests in optimization tasks to create efficient mean reverting pairs.
