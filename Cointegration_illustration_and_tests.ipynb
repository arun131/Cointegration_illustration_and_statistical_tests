```python
import numpy as np
import pandas as pd

def normalize_values(data):
    """
    Normalize the values of a numpy array, list, or pandas Series between 0 and 1.

    Args:
    - data (numpy.ndarray, list, pandas.Series): Input data to be normalized.

    Returns:
    - numpy.ndarray, pandas.Series: Normalized data.
    """

    # Convert data to numpy array if it's not already
    if not isinstance(data, np.ndarray):
        data = np.array(data)

    # Calculate the minimum and maximum values
    min_val = np.min(data)
    max_val = np.max(data)

    # Normalize the data between 0 and 1
    normalized_data = (data - min_val) / (max_val - min_val)

    # If input is pandas Series, return a Series with the same index
    if isinstance(data, pd.Series):
        normalized_data = pd.Series(normalized_data, index=data.index)

    return normalized_data
```

### Not Cointegrated

```python
m1_stock_a, m1_stock_b = not_cointegrated(plot = True)
```

```python
##### WITHOUT NORMALIZATION
# Perform cointegration tests
eg_res, jon_res, po_res, adf_res = determine_cointegration(m1_stock_a, m1_stock_b)

# Estimation
print("Estimation without Normalization")
print("Engle-Granger test:", "Cointegration detected" if eg_res else "No cointegration")

# Johansen test
print("Johansen test:", "Cointegration detected" if jon_res else "No cointegration")

# Phillips-Ouliaris test
print("Phillips-Ouliaris test:", "Cointegration detected" if po_res else "No cointegration")

# ADF-GLS test
print("ADF-GLS test:", "Cointegration detected" if adf_res else "No cointegration")
```

```python
##### AFTER NORMALIZATION
# Perform cointegration tests
eg_res, jon_res, po_res, adf_res = determine_cointegration(normalize_values(m1_stock_a), normalize_values(m1_stock_b))

# Estimation
print("Estimation with Normalization")
print("Engle-Granger test:", "Cointegration detected" if eg_res else "No cointegration")

# Johansen test
print("Johansen test:", "Cointegration detected" if jon_res else "No cointegration")

# Phillips-Ouliaris test
print("Phillips-Ouliaris test:", "Cointegration detected" if po_res else "No cointegration")

# ADF-GLS test
print("ADF-GLS test:", "Cointegration detected" if adf_res else "No cointegration")
```
